##### So here we can say that it our data is linear distributed so we can apply Linear Reggression model here

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing  import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_feature=x.select_dtypes(["int","float"]).columns
```

```python
num_feature
```

```python
cat_feature=x.select_dtypes("object").columns
```

```python
cat_feature
```

```python
num_pipeline=Pipeline([
    ("simple_imputer", SimpleImputer(strategy="median")),
    ("standard_scaler",StandardScaler())
])

cat_pipeline=Pipeline([
    ("simple_imputer",SimpleImputer(strategy="most_frequent")),
    ("encoder",OneHotEncoder(sparse_output=False))
])

final_pipeline=ColumnTransformer([
    ("num_pipeline",num_pipeline,num_feature),
    ("cat_pipeline",cat_pipeline,cat_feature)
])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

final_pipeline.fit(x_train)
x_train_tr=final_pipeline.transform(x_train)
x_test_tr=final_pipeline.transform(x_test)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train_tr,y_train)
```

```python
y_pred_train=model.predict(x_train_tr)
y_pred_test=model.predict(x_test_tr)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("The train error is :", mean_squared_error(y_pred_train,y_train))
print("The Testing error is :",mean_squared_error(y_pred_test,y_test))
```

```python
print ("the training accuracy is :", r2_score(y_pred_train,y_train))
print("the testing accuracy is:", r2_score(y_pred_test,y_test))
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model=SGDRegressor(max_iter=1000,eta0=0.05)
```

```python
model.fit(x_train_tr,y_train)
y_pred_train=model.predict(x_train_tr)
y_pred_test=model.predict(x_test_tr)
print("The train error is :", mean_squared_error(y_pred_train,y_train))
print("The Testing error is :",mean_squared_error(y_pred_test,y_test))
print ("the training accuracy is :", r2_score(y_pred_train,y_train))
print("the testing accuracy is:", r2_score(y_pred_test,y_test))
```
