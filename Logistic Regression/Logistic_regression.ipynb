```python
splitter=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=123)
for train_index,test_index in splitter.split(x,y):
    x_train=x.iloc[train_index]
    x_test=x.iloc[test_index]
    y_train=y.iloc[train_index]
    y_test=y.iloc[test_index]
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
probability=model.predict(x_test)
```

```python
(y_test==probability).sum()/len(y) *100
```

```python
print(accuracy_score(y_test,probability))
```

```python
print(classification_report(y_test,probability))
```

```python
print(confusion_matrix(y_test,probability))
```

```python
log=LogisticRegression()
log.fit(x_train,y_train)
y_hat_train=log.predict(x_train)
y_hat_test=log.predict(x_test)
training_accuracy=accuracy_score(y_hat_train,y_train)
testing_accuracy=accuracy_score(y_hat_test,y_test)
print("Logstic model")
print(f"Training Accuracy : {training_accuracy}")
print(f"Testing Accuracy : {testing_accuracy}")
print(f"Training Confusion matrix :\n {confusion_matrix(y_hat_train,y_train)}")
print(f"Testing Confusion matrix :\n {confusion_matrix(y_hat_test,y_test)}")
print(f"Training classification report :\n {classification_report(y_hat_train,y_train)}")
print(f"Testing Classfication report:\n {classification_report(y_hat_test,y_test)}")
```
